#k-neaarest neighbour alrgorthim is a supervised machine learning algorithm that is a classified technique that train the data by finding the distance from each point to the required point .

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/Cancer_Data.csv")
df
```

```python
df.info()
```

```python
df=df.drop(['id','Unnamed: 32'],axis=1)
df['diagnosis']=df['diagnosis'].replace({'M':1,'B':0})
df
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
reg = KNeighborsClassifier() # default k value = 5
reg.fit(X_train , y_train)
```

**our data training is completed here the algorithm stores the values and and does the mathematical calculation when we do the predictions**


#Train performance
- to check the performace of the model we need to check the three categories.
1. confusion matrix
2. accuracy score

#confusion matrix- this is a measure that is used to analyse the classification performance
- when the model gives the output then it has 4possible ways.
- actual 1-predicted 1=tp
- actuall 0-predicted 0=tn
- actual 1-predicted 0=fn
- actual 0-predicted 1=fp


```python
pred_answer=reg.predict(X_train)
```

```python
training_data=pd.DataFrame()

training_data['answers']=y_train
training_data['predictions']=pred_answer
training_data
```

```python
X_train.shape
```

```python
training_data['answers'].value_counts()
#actual ansewes
```

```python
training_data['predictions'].value_counts()
```

#implementing confusion matrinx with out fn means using loops

```python
tp=0
tn=0
fp=0
fn=0
for i in training_data.index:
  if training_data['answers'][i]==0 and training_data['predictions'][i]==0:
    tn=tn+1
  elif training_data['answers'][i]==1 and training_data['predictions'][i]==1:
     tp=tp+1
  elif training_data['answers'][i]==0 and training_data['predictions'][i]==1:
      fp=fp+1
  else:
      fn=fn+1
```

```python
print(f"True positive {tp}")
print(f"True negative {tn}")
print(f"False positive {fp}")
print(f"False negative {fn}")
```

```python
#with the built in fn
print(f"confusion matrix :{confusion_matrix(y_train,pred_answer)}")
```

# Accuracy - out of total positive outcomes how many are actually positive .
- this can be find using the values got buy the confusion matrix or by the built in fn called accuracy score.

```python
accuracy=(tp+tn)/(tp+tn+fp+fn)
```

```python
print(f"The accuract of the data is {accuracy}")
```

```python
#with fn
print(f"accuracy score {accuracy_score(y_train,pred_answer)}")
```

# calssification report
- this is the evaluation metrics or the model metrics for the classification so that we can know that how well our model is trained.
1. precision- out of total postive outcomes how many are actually correct is know as precision
- precision focuses on the predicted positive.
formula = tp/tp+fp
          tn/tn+fn
2. recall- focuses on the actual positive values.
formula=tp/tp+fn
3. f1-score- this metric is used to combine the precision and recall in a single vale if balances the model performance.
 formula= 2*p*r/p+r
 4. support- means no.of actual occurance of the class.


```python
#with fn
print(f"Classification report {classification_report(y_train,pred_answer)}")
```

# how to find the best k-value.
- when the accuracy of the data more for a particular k value then we can say that is a perfect value for our model.


```python
k_value=np.arange(3,51,2)
li=[]
for i in k_value:
  mod=KNeighborsClassifier(n_neighbors=i)
  mod.fit(X_train,y_train)
  li.append(accuracy_score(y_train,mod.predict(X_train)))
li

```

**we want to get the max value of the accuracy and for that we need the k value.**

```python
print(k_value[li.index(max(li))])
```
